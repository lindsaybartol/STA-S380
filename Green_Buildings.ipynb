```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = df[df['leasing_rate'] > 10]
len(df)
df.describe()
```

```python
# story ranges
df_1_to_5 = df[(df['stories'] >= 1) & (df['stories'] <= 5)]
df_5_to_10 = df[(df['stories'] > 5) & (df['stories'] <= 10)]
df_10_to_20 = df[(df['stories'] > 10) & (df['stories'] <= 20)]
df_20_plus = df[df['stories'] > 20]

#plot
plt.figure(figsize=(14, 12))

# Plot 1-5 stories
plt.subplot(2, 2, 1)
sns.boxplot(x='green_rating', y='Rent', data=df_1_to_5)
plt.title('Rent vs Green Status (1-5 Stories)')
plt.xlabel('Green Certification (1 = Green Certified, 0 = Not Green Certified)')
plt.ylabel('Rent (per square foot per year)')

# Plot 5-10 stories
plt.subplot(2, 2, 2)
sns.boxplot(x='green_rating', y='Rent', data=df_5_to_10)
plt.title('Rent vs Green Status (5-10 Stories)')
plt.xlabel('Green Certification (1 = Green Certified, 0 = Not Green Certified)')
plt.ylabel('Rent (per square foot per year)')

# Plot 10-20 stories
plt.subplot(2, 2, 3)
sns.boxplot(x='green_rating', y='Rent', data=df_10_to_20)
plt.title('Rent vs Green Status (10-20 Stories)')
plt.xlabel('Green Certification (1 = Green Certified, 0 = Not Green Certified)')
plt.ylabel('Rent (per square foot per year)')

# Plot 20+ stories
plt.subplot(2, 2, 4)
sns.boxplot(x='green_rating', y='Rent', data=df_20_plus)
plt.title('Rent vs Green Status (20+ Stories)')
plt.xlabel('Green Certification (1 = Green Certified, 0 = Not Green Certified)')
plt.ylabel('Rent (per square foot per year)')

plt.tight_layout()
plt.show()
```

```python
# story ranges
df_1_to_5 = df[(df['stories'] >= 1) & (df['stories'] <= 5)]
df_5_to_10 = df[(df['stories'] > 5) & (df['stories'] <= 10)]
df_10_to_20 = df[(df['stories'] > 10) & (df['stories'] <= 20)]
df_20_plus = df[df['stories'] > 20]

#median rents
medians = {
    '1-5 Stories': {
        'Green': df_1_to_5[df_1_to_5['green_rating'] == 1]['Rent'].median(),
        'Non-Green': df_1_to_5[df_1_to_5['green_rating'] == 0]['Rent'].median()
    },
    '5-10 Stories': {
        'Green': df_5_to_10[df_5_to_10['green_rating'] == 1]['Rent'].median(),
        'Non-Green': df_5_to_10[df_5_to_10['green_rating'] == 0]['Rent'].median()
    },
    '10-20 Stories': {
        'Green': df_10_to_20[df_10_to_20['green_rating'] == 1]['Rent'].median(),
        'Non-Green': df_10_to_20[df_10_to_20['green_rating'] == 0]['Rent'].median()
    },
    '20+ Stories': {
        'Green': df_20_plus[df_20_plus['green_rating'] == 1]['Rent'].median(),
        'Non-Green': df_20_plus[df_20_plus['green_rating'] == 0]['Rent'].median()
    }
}

# Convert the dictionary to a DataFrame for easier plotting
median_data = pd.DataFrame(medians).T.reset_index().melt(id_vars='index')
median_data.columns = ['Story Range', 'Green Status', 'Median Rent']

# color palette
custom_palette = {'Green': 'green', 'Non-Green': 'gray'}

#plot
plt.figure(figsize=(14, 12))

# Plot 1-5 stories
plt.subplot(2, 2, 1)
sns.barplot(x='Green Status', y='Median Rent', data=median_data[median_data['Story Range'] == '1-5 Stories'], palette=custom_palette)
plt.title('Median Rent (1-5 Stories)')
plt.ylabel('Median Rent (per square foot per year)')

# Plot 5-10 stories
plt.subplot(2, 2, 2)
sns.barplot(x='Green Status', y='Median Rent', data=median_data[median_data['Story Range'] == '5-10 Stories'], palette=custom_palette)
plt.title('Median Rent (5-10 Stories)')
plt.ylabel('Median Rent (per square foot per year)')

# Plot 10-20 stories
plt.subplot(2, 2, 3)
sns.barplot(x='Green Status', y='Median Rent', data=median_data[median_data['Story Range'] == '10-20 Stories'], palette=custom_palette)
plt.title('Median Rent (10-20 Stories)')
plt.ylabel('Median Rent (per square foot per year)')

# Plot 20+ stories
plt.subplot(2, 2, 4)
sns.barplot(x='Green Status', y='Median Rent', data=median_data[median_data['Story Range'] == '20+ Stories'], palette=custom_palette)
plt.title('Median Rent (20+ Stories)')
plt.ylabel('Median Rent (per square foot per year)')

plt.tight_layout()
plt.show()
```

```python
for story_range, values in medians.items():
    print(f"Story Range: {story_range}")
    for status, median_rent in values.items():
        print(f"  {status} Buildings: Median Rent = ${median_rent:.2f} per square foot per year")
```

```python
#let's only include buildings 10 - 25 stories, and buildings under the age of thirty
df_filtered = df[(df['age'] < 30) & (df['stories'] >= 10) & (df['stories'] <= 25)]

df_filtered
```

```python
# color palette
custom_palette = {'1': 'green', '0': 'gray'}

#box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='green_rating', y='Rent', data=df_filtered, palette=custom_palette)
plt.title('Rent Distribution by Green Certification (10-25 Stories, Under 30 Years Old)')
plt.xlabel('Green Certification (1 = Green Certified, 0 = Not Green Certified)')
plt.ylabel('Rent (per square foot per year)')
plt.show()
```

```python
mean_green = df_filtered[df_filtered['green_rating'] == 1]['Rent'].mean()
mean_non_green = df_filtered[df_filtered['green_rating'] == 0]['Rent'].mean()

# Calculate medians for green-certified and non-green-certified buildings
median_green = df_filtered[df_filtered['green_rating'] == 1]['Rent'].median()
median_non_green = df_filtered[df_filtered['green_rating'] == 0]['Rent'].median()

# Print the results
print(f"Mean Rent for Green-Certified Buildings: {mean_green:.2f} per square foot per year")
print(f"Mean Rent for Non-Green-Certified Buildings: {mean_non_green:.2f} per square foot per year")
print(f"Median Rent for Green-Certified Buildings: {median_green:.2f} per square foot per year")
print(f"Median Rent for Non-Green-Certified Buildings: {median_non_green:.2f} per square foot per year")
```

```python
#class a and b
df_class_a = df_filtered[df_filtered['class_a'] == 1]
df_class_b = df_filtered[df_filtered['class_b'] == 1]

#color palette
custom_palette = {'1': 'green', '0': 'gray'}

plt.figure(figsize=(12, 6))

#class a
plt.subplot(1, 2, 1)
sns.boxplot(x='green_rating', y='Rent', data=df_class_a, palette=custom_palette)
plt.title('Class A Buildings')
plt.xlabel('Green Certification (1 = Green Certified, 0 = Not Green Certified)')
plt.ylabel('Rent (per square foot per year)')

#plot class b
plt.subplot(1, 2, 2)
sns.boxplot(x='green_rating', y='Rent', data=df_class_b, palette=custom_palette)
plt.title('Class B Buildings')
plt.xlabel('Green Certification (1 = Green Certified, 0 = Not Green Certified)')
plt.ylabel('Rent (per square foot per year)')

plt.tight_layout()
plt.show()
```
